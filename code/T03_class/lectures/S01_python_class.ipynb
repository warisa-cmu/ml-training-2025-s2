### Round 7 - Motivation for Inheritance


````python
class Animal:
    def __init__(self, type, noise):
        self.type = type
        self.noise = noise
    
    def make_noise(self):
        print(f"Make some noise: {self.noise}")

    def run(self):
        raise NotImplementedError("Not implemeneted")

class Dog(Animal):
    def __init__(self, type, noise, fetch_item):
        super().__init__(type=type, noise=noise)
        self.fetch_item = fetch_item
    
    def run(self):
        print("Dog running")

class Cat(Animal):
    def __init__(self, type, noise, scracth_item):
        super().__init__(type, noise)
        self.scratch_item = scracth_item

dog = Dog(type="dog", noise="whoof", fetch_item="bone")
cat = Cat(type="cat", noise="meaw", scracth_item="sofa")

dog.run()
````
