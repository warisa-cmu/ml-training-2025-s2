### Round 7 - Motivation for Inheritance


````python
class Animal:
    def __init__(self, animal_type, animal_noise, animal_fetch_item, animal_scratch_item):
        self.type = animal_type
        self.noise = animal_noise
        self.fetch_item = animal_fetch_item
        self.scratch_item = animal_scratch_item
    
    def make_sound(self):
        print(self.noise)
        
    def fetch(self):
        print(self.type, self.fetch_item)
        
    def scratch(self):
        print(self.type, self.scratch_item)

dog = Animal(animal_type="dog", animal_noise="woof", animal_fetch_item="bone", animal_scratch_item="")
cat = Animal(animal_type="cat", animal_noise="meaw", animal_fetch_item="", animal_scratch_item="sofa")

dog.fetch()
cat.scratch()

dog.scratch()
````

### Round 8 - Inheritance (1st Attempt)


````python
class Animal:
    
    def make_noise(self):
        print(self.noise)
        

class Dog(Animal):
    
    def __init__(self, animal_type, animal_noise, animal_fetch_item):
        self.type = animal_type 
        self.noise = animal_noise
        self.fetch_item = animal_fetch_item
        
    def fetch(self):
        print(self.type, self.fetch_item)
        
class Cat(Animal):
    def __init__(self, animal_type, animal_noise, animal_scratch_item):
        self.type = animal_type 
        self.noise = animal_noise
        self.scratch_item = animal_scratch_item
        
    def scratch(self):
        print(self.type, self.scratch_item)
        

dog = Dog(animal_type="dog", animal_noise="woof", animal_fetch_item="bone")
cat = Cat(animal_type="cat", animal_noise="meaw", animal_scratch_item="sofa")

dog.fetch()
# dog.scratch()

cat.scratch()
# cat.fetch()
````

### Round 9 - Inheritance (Final)


````python
class Animal:
    def __init__(self, animal_type, animal_noise):
        self.type = animal_type 
        self.noise = animal_noise
    
    def make_noise(self):
        print(self.noise)
        

class Dog(Animal):
    
    def __init__(self, animal_type, animal_noise, animal_fetch_item):
        super().__init__(animal_type=animal_type, animal_noise=animal_noise)
        self.fetch_item = animal_fetch_item
        
    def fetch(self):
        print(self.type, self.fetch_item)
        
class Cat(Animal):
    def __init__(self, animal_type, animal_noise, animal_scratch_item):
        super().__init__(animal_type=animal_type,animal_noise=animal_noise)
        self.scratch_item = animal_scratch_item
        
    def scratch(self):
        print(self.type, self.scratch_item)
        

dog = Dog(animal_type="dog", animal_noise="woof", animal_fetch_item="bone")
cat = Cat(animal_type="cat", animal_noise="meaw", animal_scratch_item="sofa")

dog.fetch()
# dog.scratch()

cat.scratch()
# cat.fetch() 

dog.noise
````
