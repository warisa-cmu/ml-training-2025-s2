```python
import optuna
```

```python
def objective(trial: optuna.trial.Trial):
    x = trial.suggest_float("x", -10, 10)
    obj = (x - 2) ** 2
    return obj
```

```python
study = optuna.create_study()
study.optimize(objective, n_trials=1000)
```

```python
best_params = study.best_params
print(best_params)
```
