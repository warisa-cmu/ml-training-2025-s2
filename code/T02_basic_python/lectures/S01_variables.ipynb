$$slope = \frac{(y_2 - y_1)} { (x_2 - x_1)}$$
$$distance = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}$$


```python
# Calculate slope and distance between points (2, 2) and (6, 10)
# x1 = 2
# y1 = 2
x1, y1 = 2, 2
x2, y2 = 6, 10
slope = (y2 - y1) / (x2 - x1)
distance = ((x2 - x1) ** 2 + (y2 - y1) ** 2) ** (0.5)
print(slope, distance)
```
