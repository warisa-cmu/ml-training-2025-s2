```python
# If, Elif, Else
# Write a program that asks the user for a number and prints whether the number is positive, negative, or zero.
# num = int(input("Enter a number"))
num = 10
if type(num) == str:
    raise Exception("You enter a string")

if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")



```

```python
exps = ["exp1", "exp2", "exp3"]
params = [{"eta": 0.1}, {"eta": 0.2}, {"eta": 0.3}]
filenames = ["File1", "File2", "File3"]

for i in range(len(exps)):
    print(exps[i], params[i], filenames[i])

for exp, param, filename in zip(exps, params, filenames):
    print(exp, param, filename)
```

```python
# While Loop
# Write a program that prints numbers from 5 down to 1 using a while loop.
count = 5
while count > 0:
    print(count)
    count = count - 1
```
