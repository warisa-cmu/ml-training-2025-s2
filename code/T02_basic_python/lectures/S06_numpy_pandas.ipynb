```python
# Create a numpy array with the elements [1, 2, 3, 4, 5]
a = [1,2,3,4,5]

a_np = np.array(a)

a_np.mean()
a_np.max()
a_np.min()
a_np.std()
a_np.var()
a_np.argmin()
a_np.argmax()


print(np.argmin(a_np))
print(a_np.argmin())
```

```python
# Create a numpy array with the elements 1 to 100.
b = np.arange(start=1,stop=100,step=2)
print(b)

```

```python
# Print the shape of the array from the previous step
b.shape
```

```python
# Create a 2x2 numpy array with elements
# |1, 2|
# |3, 4|
c = [[1, 2, 3], [3, 4, 1], [5,5,5]]
d = np.array(c)
# print(d)
# print(d.shape)
print(d[0,:])
print(d[:1,:])
```

```python
# Calculate the sum of all elements in the 2x2 array
print(d)
d.sum(axis=1)
```

```python
# Calculate the mean of the elements in the 2x2 array

d.mean(axis=0)
```

```python
# Reshape the 2x2 array to a 1x4 array and print it
a = [[1,2],[3,4]]
a_np = np.array(a)
print(a_np)
b = a_np.reshape(1,4,1)
print(b)
```

# Pandas


```python
# Import the pandas library as pd
import pandas as pd
```

```python
# Create a pandas DataFrame from the following dictionary:
data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Age": [23, 25, 30, 22],
    "Score": [85, 90, 88, 95],
}

df = pd.DataFrame(data=data)
df
```

```python
# Print the first 2 rows of the DataFrame
df.head(2)
df.tail(2)
```

```python
# Print the column names of the DataFrame
df.columns
```

```python
sr = df["Score"] >= 90

sr.map({True: "Pass", False: "Failed"})
```

```python
# Add a new column called 'Passed' that is True if the 'Score' is greater than or equal to 90, otherwise False
sr = df["Score"] >= 90
sr = sr.map({True: "Pass", False: "Failed"})
df["Pass"] = sr
df
```

```python
# Calculate and print the average age of the students
df["Age"].var()
```
